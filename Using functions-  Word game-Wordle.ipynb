```python
import pandas as pd
import random
import re
from IPython.display import clear_output

def play_game_match():
    ## create a list of already used words
    try:
        used_words = pd.read_csv('usedwords.csv', header=None).T.values[0].tolist()
    except:
        used_words = []
        
    ## chose a word from list of words    
    words = pd.read_csv('word_from_bestwordlist.csv', header=None).T.values[0].tolist()
    words = [i.split(" ") for i in words]
    words = [item.lower() for sublist in words for item in sublist]
    
    ## chosing a random word that is not in the used file
    word = random.choice(list(set(words)-set(used_words))).strip().lower()
    
    ## save new used_words 
    used_words.append(word)
    pd.DataFrame(used_words).to_csv('usedwords.csv', header=None, index=None)
    
    # this list will keep the list of entry words with colors added 
    patterns = []
    
    # this dictionary will contain letters and colors corresponding
    alphabet = dict()
    for char in "abcdefghijklmnopqrstuvwxyz":
        alphabet[char] = ""
    
    # number of tries already done and number of tries left
    tries = 0
    allowed_tries =6
    
    while allowed_tries > 0:
        # get user's entry
        entry = input("Enter a word: ").strip().lower()
        # initiate a dictionary to store letters and colors
        hidden_pattern = dict(enumerate(entry))
        
        # throw errors if the user enters not 5-letter words and words that are not in dicitonary
        if len(entry) != 5:            
            print("The word must be a 5-letter word. Please try again")
            continue
            
        if entry not in words:
            print("There is no such word as " + entry + ". Please try again")
            continue
        
        # update tries
        tries +=1
        allowed_tries -=1    
            
        # clear the output area
        clear_output(wait=True)
        
        
        # see if entry has some letters matched with the word and update patterns and alphabet
        if set(entry).intersection(set("abcdefghijklmnopqrstuvwxyz"))==set(entry):
            for k,v in enumerate(entry):
                if v not in word:
                    hidden_pattern[k] = "\x1b[6;30;47m" + v + "\x1b[0m"
                    alphabet[v] = "6;30;47"
                elif v in word and dict(enumerate(word))[k]==v:
                    hidden_pattern[k]  = "\x1b[6;30;42m" + v + "\x1b[0m"
                    alphabet[v] = "6;30;42"
                elif v in word and dict(enumerate(word))[k]!=v:
                    hidden_pattern[k] = "\x1b[6;30;43m" + v + "\x1b[0m"
                    if alphabet[v]=="":
                        alphabet[v] = "6;30;43"
                       
            patterns.append(" ".join(hidden_pattern.values()))
            
            # print out the output
            for i in patterns:
                print(i)
            print("\n")
            for char, v in alphabet.items():
                if v == "":
                    print(char, end=" ")
                else:
                    print("\x1b[" + v + "m" + char + "\x1b[0m", end =" ")
            print("\n")
            
            if entry==word:
                print("You win! The word was " + word +". You won in " + str(tries) + " tries")
                break
        # because we already covered the errors in the prior if's this error will never occur    
        else:
            print("You have made a mistake in your input. Please try again")
            continue
        # finish the game if there are no more tries left
        if allowed_tries==0:
            print("Game over! The word was " + word + ".")
            break
        
```

```python
play_game_match()
```
